### Descripción del Problema: Clase GestionVenta
El objetivo es modelar un sistema que permita gestionar el ciclo de vida de una transacción comercial desde la prospección hasta el cierre. Cada venta es tratada como una oportunidad única con su propia identidad, valor económico y estado dentro del embudo de ventas.

### 1. Definición de Atributos
Los atributos representan las características esenciales que definen cada oportunidad de negocio:

* idTransaccion: Identificador único para el registro contable y seguimiento de la venta.

* valorEstimado: Valor numérico (generalmente en moneda local) que representa el monto del contrato o producto.

* clientePotencial: Nombre de la persona o empresa interesada en la adquisición.

* etapaVenta: Fase actual del proceso comercial (ej. "Prospección", "Propuesta enviada", "Negociación", "Cerrada").

### 2. Definición de Funciones (Métodos)
Las funciones definen las acciones y estrategias que el vendedor ejecuta para avanzar en el proceso:

* presentarOferta(): Realiza la exposición del producto o servicio al cliente, cambiando el estado a "Propuesta enviada".

* aplicarDescuento(): Ajusta el valor estimado del cierre para incentivar la decisión de compra, vinculando la autorización al idTransaccion.

* consultarStatus(): Devuelve un reporte del progreso de la negociación y las probabilidades de éxito actuales.

 ![texto del vínculo](vendedorinfografia.png)

```python
# Programado por Sebastian Farfan
# Tema: Gestión de Ciclo de Venta

class GestionVenta:
    def __init__(self, idTransaccion, valorEstimado, clientePotencial):
        self.idTransaccion = idTransaccion
        self.valorEstimado = valorEstimado
        self.clientePotencial = clientePotencial
        self.etapaVenta = "Prospección"

    def presentarOferta(self):
        self.etapaVenta = "Propuesta enviada"
        print(f"💼 Fase 1: Oferta presentada al cliente {self.clientePotencial} con éxito.")

    def aplicarDescuento(self):
        # Simulamos una negociación que cambia el estado
        self.etapaVenta = "Negociación"
        print(f"📉 Fase 2: Descuento aplicado a la transacción {self.idTransaccion}.")

    def consultarStatus(self):
        print("\n--- 📊 REPORTE COMERCIAL DE VENTA ---")
        print(f"ID Transacción: {self.idTransaccion}")
        print(f"Cliente Potencial: {self.clientePotencial}")
        print(f"Valor del Trato: ${self.valorEstimado}")
        print(f"Etapa Actual: {self.etapaVenta.upper()}")
        print("-------------------------------------")

# --- FLUJO INTERACTIVO LINEAL ---

print("💰 REGISTRO DE NUEVA OPORTUNIDAD")
id_t = input("Escriba el ID de la transacción: ")
valor = input("Escriba el valor estimado del contrato: ")
cliente = input("Escriba el nombre del cliente potencial: ")

# Creación del objeto basado en la clase Vendedor
venta = GestionVenta(id_t, valor, cliente)

# Ejecución de pasos comerciales
input("\nPulse Enter para PRESENTAR OFERTA...")
venta.presentarOferta()

input("Pulse Enter para INICIAR NEGOCIACIÓN (Descuento)...")
venta.aplicarDescuento()

input("Pulse Enter para CONSULTAR STATUS FINAL...")
venta.consultarStatus()
```
